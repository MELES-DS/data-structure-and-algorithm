```python
def insertion_sort(arr):
	i = 1
	while i < len(arr):
		key = arr[i]
		j = i - 1
		while j >= 0 and key < arr[j]:
			arr[j + 1] = arr[j]
			j -= 1
		arr[j + 1] = key
		i += 1

# Example usage
arr = [12, 34, 11, 13, 5, 554, 45, 6, 78]
insertion_sort(arr)
print("Sorted array is:", arr)

# Method 2 :
def insertion_sort_recursive(arr, n):
	if n <= 1:
		return
	insertion_sort_recursive(arr, n-1)
	last = arr[n-1]
	j = n-2
	while j >= 0 and arr[j] > last:
		arr[j + 1] = arr[j]
		j -= 1
	arr[j + 1] = last

# Example usage
arr = [12, 11, 13, 5, 6, 2, 3, 23, 56, 78, 90]
insertion_sort_recursive(arr, len(arr))
print("Sorted array is:", arr)

# 2. Selection Sort
# The selection_sort function is already defined and used below.
# No additional code needed here.
def selection_sort(arr):
	n = len(arr)
	i = 0
	while i < n:
		min_idx = i
		j = i + 1
		while j < n:
			if arr[j] < arr[min_idx]:
				min_idx = j
			j += 1
		arr[i], arr[min_idx] = arr[min_idx], arr[i]
		i += 1

# Example usage
arr = [64, 25, 12, 22, 11]
selection_sort(arr)
print("Sorted array is:", arr)
```
