```python
import networkx as nx
import matplotlib.pyplot as plt

class Node:  
    def __init__(self, key):
        self.left = None  
        self.right = None  
        self.val = key  

class BinarySearchTree:
    def __init__(self):
        self.root = None 
        self.tree = nx.DiGraph() 

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
            self.tree.add_node(key)  
        else:
            self._insert_rec(self.root, key)
    
    def _insert_rec(self, current_node, key):
        if key < current_node.val:
            if current_node.left is None:
                current_node.left = Node(key)
                self.tree.add_node(key)
                self.tree.add_edge(current_node.val, key)  
            else:
                self._insert_rec(current_node.left, key)
        else:
            if current_node.right is None:
                current_node.right = Node(key)
                self.tree.add_node(key)
                self.tree.add_edge(current_node.val, key)  
            else:
                self._insert_rec(current_node.right, key)

    def visualize(self):
        """Visualize the binary tree using matplotlib."""
        pos = self._layout()  # Use a predefined layout
        nx.draw(self.tree, pos, with_labels=True, arrows=True)
        plt.title("Binary Tree Visualization")
        plt.show()
    
    def _layout(self):
        """Create a layout for the nodes to visualize the tree."""
        pos = {}
        self._layout_rec(self.root, pos, 0, 0, 1)  # Start layout from the root
        return pos
    
    def _layout_rec(self, node, pos, x, y, layer):
        """ Recursive function to layout the nodes."""
        if node is not None:
            pos[node.val] = (x, y)  # Position of the current node
            self._layout_rec(node.left, pos, x - 1 / (2 ** layer), y - 1, layer + 1)  # Left child
            self._layout_rec(node.right, pos, x + 1 / (2 ** layer), y - 1, layer + 1)  # Right child
    
    def inorder_traversal(self):
        stack = []
        current = self.root
        result = []

        while stack or current:
            while current:
                stack.append(current)
                current = current.left

            current = stack.pop()
            result.append(current.val)
            current = current.right

        return result
    
    def preorder_traversal(self):
        if not self.root:
            return []

        stack = [self.root]
        result = []

        while stack:
            current = stack.pop()
            result.append(current.val)

            # Push right first so that left is processed first
            if current.right:
                stack.append(current.right)
            if current.left:
                stack.append(current.left)

        return result
    
    def postorder_traversal(self):
        if not self.root:
            return []

        stack1 = [self.root]
        stack2 = []
        result = []

        while stack1:
            current = stack1.pop()
            stack2.append(current)

            if current.left:
                stack1.append(current.left)
            if current.right:
                stack1.append(current.right)

        while stack2:
            result.append(stack2.pop().val)

        return result


BSearchTree = BinarySearchTree()
BSearchTree.insert(20)
BSearchTree.insert(50)
BSearchTree.insert(5)
BSearchTree.insert(13)
BSearchTree.insert(17)
BSearchTree.insert(12)
BSearchTree.insert(28)
BSearchTree.insert(4)
BSearchTree.insert(1)
BSearchTree.insert(74)

BSearchTree.visualize()  
print(f" In Order Traversal output {BSearchTree.inorder_traversal()}")
print(f" Pre Order Traversal output {BSearchTree.preorder_traversal()}")
print(f" Post Order Traversal output {BSearchTree.postorder_traversal()}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BT:
    def __init__(self):
        self.root = None
        self.tree = nx.Graph()

    def insert_node(self, data):
        new_node = Node(data)
        if self.root is None:
            self.root = new_node
            self.tree.add_node(data)
            return

        queue = [self.root]
        while queue:
            current = queue.pop(0)
            if current.left is None:
                current.left = new_node
                self.tree.add_node(data)
                self.tree.add_edge(current.data, data)  # Use current.data
                return
            else:
                queue.append(current.left)

            if current.right is None:
                current.right = new_node
                self.tree.add_node(data)
                self.tree.add_edge(current.data, data)  # Use current.data
                return
            else:
                queue.append(current.right)

    def inorder_traversal(self):
        stack = []
        current = self.root
        result = []

        while stack or current:
            while current:
                stack.append(current)
                current = current.left

            current = stack.pop()
            result.append(current.data)
            current = current.right

        return result
    
    def preorder_traversal(self):
        if not self.root:
            return []

        stack = [self.root]
        result = []

        while stack:
            current = stack.pop()
            result.append(current.data)

            # Push right first so that left is processed first
            if current.right:
                stack.append(current.right)
            if current.left:
                stack.append(current.left)

        return result
    
    def postorder_traversal(self):
        if not self.root:
            return []

        stack1 = [self.root]
        stack2 = []
        result = []

        while stack1:
            current = stack1.pop()
            stack2.append(current)

            if current.left:
                stack1.append(current.left)
            if current.right:
                stack1.append(current.right)

        while stack2:
            result.append(stack2.pop().data)

        return result
    
    def visualize(self):
        """Visualize the binary tree using matplotlib."""
        pos = self._layout()  # Use a predefined layout
        
        nx.draw(self.tree, pos, with_labels=True, arrows=True)
        plt.title("Binary Tree Visualization")
        plt.show()

    def _layout(self):
        """Create a layout for the nodes to visualize the tree."""
        pos = {}
        self._layout_rec(self.root, pos, 0, 0, 1)  # Start layout from the root
        return pos

    def _layout_rec(self, node, pos, x, y, layer):
        """Recursive function to layout the nodes."""
        if node is not None:
            pos[node.data] = (x, y)  # Position of the current node
            self._layout_rec(node.left, pos, x - 1 / (2 ** layer), y - 1, layer + 1)  # Left child
            self._layout_rec(node.right, pos, x + 1 / (2 ** layer), y - 1, layer + 1)  # Right child


tree = BT()
tree.insert_node(1)
tree.insert_node(2)
tree.insert_node(3)
tree.insert_node(4)
tree.insert_node(5)
print("Visualize Tree")
tree.visualize()

print(f" In Order Traversal output {tree.inorder_traversal()}")
print(f" Pre Order Traversal output {tree.preorder_traversal()}")
print(f" Post Order Traversal output {tree.postorder_traversal()}")
```
